```python
# %% Convert PDF to text and add to context
from PyPDF2 import PdfReader
import os


def pdf_to_text(pdf_path, output_txt):
    reader = PdfReader(pdf_path)
    text = ""
    for i, page in enumerate(reader.pages):
        text += f"\n--- Страница {i + 1} ---\n"
        text += page.extract_text()

    with open(output_txt, "w", encoding="utf-8") as f:
        f.write(text)
    print(f"Текст извлечен и сохранён в {output_txt}")


# Укажите путь к вашему PDF
pdf_path = "Context/Automatic_Parkinsons_disease_detection_from_speech_Layer_selection_vs_adaptation_of_foundation_models.pdf"  # ← измените путь при необходимости
txt_path = "Context/Automatic_Parkinsons_disease_detection_from_speech_Layer_selection_vs_adaptation_of_foundation_models.txt"

# Проверка существования
if os.path.exists(pdf_path):
    pdf_to_text(pdf_path, txt_path)
else:
    print("Файл не найден:", pdf_path)
    print("Текущая директория:", os.getcwd())
```
